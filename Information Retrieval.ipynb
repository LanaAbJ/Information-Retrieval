```python
import os
import string
import nltk
import string
import pandas as pd
import numpy as np
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download("punkt")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Uploading docs into the corpus
corpus = []
directory = r'/content/drive/MyDrive/Colab Notebooks/ALL Documents'
for filename in os.listdir(directory): # listdir() is used to see what files are in the folder
  if filename.endswith('.txt'):
    with open(os.path.join(directory, filename), 'r') as f:
      contents = f.read()
      corpus.append(contents)


#####################################################

def preprocess_text(text):

    #lower casing
    lower = text.lower()
    # Tokenise words while ignoring punctuation
    tokeniser = RegexpTokenizer(r'\w+')
    tokens = tokeniser.tokenize(lower)
    #remove stopwords
    withoutStop = [token for token in tokens if token not in stopwords.words('english')]
    #stemming
    stemmer = nltk.PorterStemmer()
    stem =[stemmer.stem(nostop) for nostop in withoutStop]
  
    #lemmatise 
    lemmatiser = WordNetLemmatizer()
    lemmas = [lemmatiser.lemmatize(st) for st in stem]
    listToString = ' '.join(map(str, lemmas ))
    return listToString


# Printing the corpus before Pre-Processing
print("#########Corpus before Pre-Processing#########")
for i, contents in enumerate(corpus):
      print(f'Doc {i+1}:\n{contents}\n')
Doc1,Doc2,Doc3,Doc4,Doc5,Doc6,Doc7,Doc8,Doc9,Doc10,Doc11,Doc12 = corpus
D1 = preprocess_text(Doc1)
D2 = preprocess_text(Doc2)
D3 = preprocess_text(Doc3)
D4 = preprocess_text(Doc4)
D5 = preprocess_text(Doc5)
D6 = preprocess_text(Doc6)
D7 = preprocess_text(Doc7)
D8 = preprocess_text(Doc8)
D9 = preprocess_text(Doc9)
D10= preprocess_text(Doc10)
D11 = preprocess_text(Doc11)
D12 = preprocess_text(Doc12)

# The newer corpus after Pre-Processing
print("#########Corpus after Pre-Processing#########")
new_corpus = [D1,D2,D3,D4,D5,D6,D7,D8,D9,D10,D11,D12]
for i, contents in enumerate(new_corpus):
    print(f'Doc {i+1}:\n{contents}\n') 
```

```python
# Tf-Idf 

# Create a dataframe
TF_IDF_matrix = pd.DataFrame(new_corpus, columns=['text'])
# Create an instance of TfidfVectorizer
vectoriser = TfidfVectorizer(max_features=1000)
# Fit to the data and transform to feature matrix
TF_IDF_matrix = vectoriser.fit_transform(TF_IDF_matrix['text'])
# Convert sparse matrix to dataframe
matrixframe = pd.DataFrame.sparse.from_spmatrix(TF_IDF_matrix)
# Save mapping on which index refers to which words
col_map = {v:k for k, v in vectoriser.vocabulary_.items()}
# Rename each column using the mapping
for col in matrixframe.columns:
    matrixframe.rename(columns={col: col_map[col]}, inplace=True)
# Add row names to the dataframe
matrixframe.index = ["Doc" + str(i+1) for i in range(len(matrixframe))]

matrixframe
```

```python
new_query = input("Enter your query please :")
# Compute TF-IDF matrix for a new query
new_query_vector = vectoriser.transform([new_query])
# Create a list conataining the results of the Cosine Similarity between query and Corpus
cosinesim = []
for i in TF_IDF_matrix:
  cossim = cosine_similarity(i,new_query_vector)
  cosinesim.append(np.round(cossim[0][0],3))
# Print out the results
for i,doc in enumerate(cosinesim):
  print(f'doc{i+1} cosine similarity with {new_query}: {doc}\n')
```

```python
topdocs = {}
docNamelist = []
for i,result in enumerate(cosinesim):
  if result > 0.003:
    print(f'doc{i+1} is relevant. \n')
    for j,doc in enumerate(corpus):
      if i == j:
        topdocs[result] = doc
        docNamelist.append(j+1)
  else:
    print(f'doc{i+1} is irrelevant. \n')

thetopdocs = sorted(topdocs)
sorted_topdocs= {key:topdocs[key] for key in thetopdocs}
print("########## Top relevant docs ##########")
for result,name in zip(sorted_topdocs,docNamelist):
    print(f'Doc {name}:\n{sorted_topdocs[result]}\n') 
   
```
